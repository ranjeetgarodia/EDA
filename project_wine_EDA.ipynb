```python
# trying to find out bad,avg, good quality of wine for given dataset
#above shows quality range between 3 to 8

# very_bad_quality less than or equal to 3
#Bad_quality bewteen 3 to 5 (excluding <5)
# Avg_quality is 5 or 6 , but less than 7
# good quality between 7 &8
#very good quality >8

#print(wine_data_fromCSVFile.shape[0])

bad_quality = wine_data_fromCSVFile.loc[(wine_data_fromCSVFile.quality < 5)]
bad_quality['ratings_']='Bad'
#print(bad_quality.head(5))
len_bad_quality=len(bad_quality)
#print('bad_quality count:',len(bad_quality))

avg_quality = wine_data_fromCSVFile.loc[( (wine_data_fromCSVFile.quality >= 5) & (wine_data_fromCSVFile.quality < 7) ) ]
avg_quality['ratings_']='Avg'

good_quality = wine_data_fromCSVFile.loc[(wine_data_fromCSVFile.quality >= 7) & (wine_data_fromCSVFile.quality <=8)  ]
good_quality['ratings_']='Good'

quality_above_6 = wine_data_fromCSVFile.loc[(wine_data_fromCSVFile['quality'] > 6)]
n_above_6 = quality_above_6.shape[0]
print(n_above_6)

# display(np.round(wine_data_fromCSVFile.describe()))
quality_series=pd.Series([len(bad_quality),len(avg_quality),len(good_quality)],
                         index =['bad_quality 0-4','avg_quality 5-6','good_quality 7-8']
                                 )
quality_df=pd.DataFrame({'quality_count': quality_series})

quality_df['percentage']=quality_df.quality_count.apply(lambda x:x/total *100)
quality_df
bad_quality
```
